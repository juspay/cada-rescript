````python
def get_codeBlock(lines):  
    
    end,code = 0,"" 
    for line in lines:
        end = line['destination']
        code += line['line']
    
    return (end,code)

changes = dict()

for files in result['diffs']:
    filename = files['source']['components'][-1]
    for k in files['hunks']:
        for seg in k['segments']:
            if seg['type'] == 'ADDED':
                code = get_codeBlock(seg['lines'])

                if changes.get(filename,"") == "": 
                   changes[filename] = [code]
                else:
                   changes[filename].append(code)
````

````python
component_wise_sys_prompt = """
You are a QA automation assistant. You will receive a single frontend function or component. Your task is to generate precise, verifiable test cases in JSON format.

If the input is a function:

Treat it as developer-level logic.
Generate test cases that call the function with different arguments.
Describe input/output behavior clearly.
Focus on edge cases, expected return values, and logical correctness.

If the input is a component:

Treat it as a user-facing UI component.
Generate UI-level test cases from a user's perspective.
Describe realistic steps like clicking buttons, typing in inputs, or viewing state changes.
Focus on visible behavior and interaction.

Output Format:
Only return a valid JSON array like this:
[
  {
    "testCaseId": "TC001",
    "description": "Brief summary of what the test verifies",
    "stepsToReproduce": [
      "Step 1",
      "Step 2"
    ],
    "expectedResult": "Expected behavior"
  }
]

Do not explain or wrap the JSON in markdown
Only return JSON
Ignore deleted items or unrelated helpers
Ensure each test case is clear, unique, and easy to execute manually or with automation

"""

component_wise_prompt = """
You are given a single frontend function or component from a pull request. Based on what it does, generate appropriate test cases in JSON format.
[
  {{
    "testCaseId": "TC001",
    "description": "Brief summary of what the test verifies",
    "stepsToReproduce": [
      "Step 1",
      "Step 2",
      "..."
    ],
    "expectedResult": "Expected behavior"
  }}
]
Guidelines:

If the input is a function:
Write developer-level test cases.
Focus on how the function behaves with different inputs.
No UI steps — just logical steps like calling the function with X and expecting Y.

If the input is a component:

Write user-level test cases.
Describe how a user would interact with the UI (clicking, typing, navigating).
Test realistic behavior visible to the user.
Only return valid JSON — do not include any explanation or extra text.

Input:
{component_or_function_code}
"""
````

````python
import re
def parse(text: str) -> dict:
    cleaned = re.sub(r"```(?:json)?", "", text).strip()
    
    # Try parsing it
    try:
        parsed = json.loads(cleaned)
        return parsed
    except json.JSONDecodeError as e:
        print("❌ Failed to parse JSON:", e)
        print(text)
        return None
````

````python
import tiktoken

def get_token_count(text, model="gpt-4o"):
    encoding = tiktoken.encoding_for_model(model)

    tokens = encoding.encode(text)
    return len(tokens)
````

````python
import os
from google import genai
from google.genai import types


def generate(text):
    overall_response = ""
    client = genai.Client(
        api_key=""
    )

    model = "gemini-2.5-pro-preview-05-06"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(text=text),
            ],
        ),
    ]
    generate_content_config = types.GenerateContentConfig(
        response_mime_type="text/plain",
    )

    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        print(chunk.text, end="")
        overall_response += chunk.text
    
    return overall_response
````

````python
import os 
from jaf.types import Query 

def call_llm(q: Query):
    tkn_cnt = get_token_count(q.prompt)
    print("tokenCount: ",tkn_cnt)

    if tkn_cnt >= 120000:
        print("Using Gemini")
        resp = generate(q.prompt)
        parsed_resp = parse(resp)
    else:
        print("Using Azure")
        llm(q)
        parsed_resp = parse(q.response)

    return parsed_resp
````

````python
q = Query()

module_list = [] 
component_name = []
component_type = []
component_code = [] 
modified_code = [] 
test_cases = []

for res in result:
    moduleName = os.path.basename(res.moduleName)
    tcases = []
    for fun in res.addedFunctions:
        name,code = fun 
        content = name + "\n" + code + "\n" 
        q.prompt = component_wise_prompt.format(component_or_function_code=content)
        q.system_prompt = component_wise_sys_prompt
        resp = call_llm(q)

        if resp is None: 
            print(moduleName)
            print("Name: ",name)

        else:
            module_list.append(moduleName)
            component_name.append(name)
            component_code.append(code)
            modified_code.append(None)
            component_type.append("ADDED")
            test_cases.append(resp)
        tcases.append(resp) 
    
    for fun in res.deletedFunctions:
        name,code = fun
        content = name + "\n" + code + "\n" 
        q.prompt = component_wise_prompt.format(component_or_function_code=content)
        q.system_prompt = component_wise_sys_prompt
        resp = call_llm(q)

        if resp is None: 
            print(moduleName)
            print("Name: ",name)

        else:
            module_list.append(moduleName)
            component_name.append(name)
            component_code.append(code)
            modified_code.append(None)
            component_type.append("DELETED")
            test_cases.append(resp)
        tcases.append(resp) 

    for fun in res.modifiedFunctions:
        name,old_code,new_code,metadata = fun 
        start = metadata['new_start'][0] 
        end = metadata['new_end'][0]

        added_pieces = ""
        for (pos,code) in changes[moduleName]:
            if pos >= start and pos <= end:
                added_pieces += code

        content = name + "\n" + new_code + "\ngit diff added/modified pieces " + added_pieces
                
        q.prompt = component_wise_prompt.format(component_or_function_code=content)
        q.system_prompt = component_wise_sys_prompt
        resp = call_llm(q)

        if resp is None: 
            print(moduleName)
            print("Name: ",name)

        else:
            module_list.append(moduleName)
            component_name.append(name)
            component_code.append(new_code)
            modified_code.append(added_pieces)
            component_type.append("MODIFIED")
            test_cases.append(resp)
        tcases.append(resp) 
    
    res.testCases = tcases
````

````python
import pandas as pd 

data = {"file_name": module_list, "component_name": component_name, "component_type": component_type, "component_code": component_code, "diff_snippet": modified_code, "test_cases": test_cases}

df = pd.DataFrame(data)
````
